```python
import pandas as pd
```

```python
data=pd.read_csv('tested.csv')
data
```

```python
data.isnull().sum()
```

```python
data.drop('Cabin',axis=1,inplace=True)
data.head()
```

```python
data.insert(3,'New',data.loc[:,'Name'].str.split(',').str[1].str.split('.').str[0])
data.groupby('New')['Age'].apply(lambda x: x.isnull().sum())
```

```python
master_ind=[]
for i in data.iterrows():
    if 'Master' in i[1]['Name']:
        master_ind.append(i[0])
```

```python
mister_ind=[]
for i in data.iterrows():
    if 'Mr' in i[1]['Name']:
        mister_ind.append(i[0])
```

```python
ms_ind=[]
for i in data.iterrows():
    if 'Miss' in i[1]['Name'] or 'Ms' in i[1]['Name']:
        ms_ind.append(i[0])
```

```python
mrs_ind=[]
for i in data.iterrows():
    if 'Mrs' in i[1]['Name']:
        mrs_ind.append(i[0])
```

```python
master_df=data.loc[master_ind,['Age']]
data.loc[master_ind, 'Age'] = master_df['Age'].fillna(master_df['Age'].mean())
mister_df=data.loc[mister_ind,['Age']]
data.loc[mister_ind, 'Age'] = mister_df['Age'].fillna(mister_df['Age'].mean())
ms_df=data.loc[ms_ind,['Age']]
data.loc[ms_ind, 'Age'] = ms_df['Age'].fillna(ms_df['Age'].mean())
mrs_df=data.loc[mrs_ind,['Age']]
data.loc[mrs_ind, 'Age'] = mrs_df['Age'].fillna(mrs_df['Age'].mean())
data.isnull().sum()
```
