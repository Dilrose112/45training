```python
import pandas as pd
```

```python
data=pd.read_csv('2023_Countries by Population.csv')
data.head()
```

```python
#calculating 2025 population
data.insert(6,'pop2025',data.loc[:,'pop2023']+((data.loc[:,'pop2023']*data.loc[:,'growthRate'])*2))
data.head()
data.isnull().sum()
```

```python
#calculating the missing values in net change column
a= [(data.loc[:,'pop2000']- data.loc[:,'pop1980'])/20,(data.loc[:,'pop2010']- data.loc[:,'pop2000'])/10,(data.loc[:,'pop2022']- data.loc[:,'pop2010'])/12,(data.loc[:,'pop2023']- data.loc[:,'pop2022']),(data.loc[:,'pop2030']- data.loc[:,'pop2023'])/7,(data.loc[:,'pop2050']- data.loc[:,'pop2030'])/20]
data.loc[:,'netChange'].fillna((sum(a)/len(a))/1000,inplace=True)
data.isnull().sum()
```

```python
#calculating the missing values in world percentage
sum=data['pop2023'].sum()
data.loc[:,'worldPercentage'].fillna(data.loc[:,'pop2023']/sum,inplace=True)
data.isnull().sum()
```
